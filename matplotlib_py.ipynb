#Question1 Write a Python program to draw a line using given axis values taken
#from a text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
text_txt = {"X":[1,2,3],"Y":[2,4,1]}
pd.DataFrame(text_txt)
```

```python
plt.plot(text_txt["X"],text_txt["Y"])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("linear graph")
plt.show()
```
